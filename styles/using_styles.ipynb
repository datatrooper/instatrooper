```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.reset_defaults()

plt.style.use('https://raw.githubusercontent.com/datatrooper/instatrooper/master/styles/bluetrooper.mplstyle')
```

```python
tips = sns.load_dataset("tips")
fig = plt.figure()
ax = sns.barplot(x="day", y="total_bill", data=tips)
ax.set_title("Best rated movies")
ax.set_xlabel("Average of votes")
ax.set_ylabel("Movie titles")
plt.show()
fig.savefig('whatever.png', facecolor=fig.get_facecolor(),bbox_inches="tight", pad_inches=2)
```

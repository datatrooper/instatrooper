```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.reset_defaults()

plt.style.use('https://raw.githubusercontent.com/datatrooper/instatrooper/master/styles/bluetrooper.mplstyle')
```

```python
datatrooper_palette = [
    "#e6d3ad",
    "#967B4F",
    "#46493D",
    "#839DA5",
    "#9FADAD",
    "#2B4144",
    "#986E37"
]
```

```python
titanic = sns.load_dataset("titanic")

t_gr = pd.DataFrame(titanic.groupby("embarked").count()["survived"]).reset_index()

display(t_gr)

fig = plt.figure()
ax = sns.barplot(x = "survived", y = "embarked", data=t_gr, orient="h")
plt.grid(False)
plt.title("Países con mayor incremento"
           "\n"  # Newline: the backslash is interpreted as usual
           "en el porcentaje de área forestal")
ax.set_xlabel("")
ax.set_ylabel("")

plt.xticks([])
plt.box(False)


# jugar con estos valores
vertical_offset = 50
horizontal_offset = 0.1

for index, row in t_gr.iterrows():
    ax.text(row["survived"] + vertical_offset,index+horizontal_offset, # posicion
            row["survived"], # texto
            color='white', 
            ha="center", 
            size=35)


plt.savefig('match.png',bbox_inches='tight', transparent="True", pad_inches = 0)
plt.show()
```

```python
def ismain(row):
    if ((row["ch1"] in chars) & (row["ch2"] in chars)):
        return True
    else:
        return False
```

```python
with open('../../../data/friends/friends.txt', 'r') as f:
    lines = [line.strip() for line in f]

rels = [l.split() for l in lines if "#" not in l]
all_df = pd.DataFrame(rels, columns = ["ch1", "ch2"])
rels_df = all_df.copy()
```

```python
chars = ["Chandler", "Joey", "Ross", "Monica", "Rachel", "Phoebe"]
rels_df["ismain"] = rels_df.apply(ismain, axis=1)
rels_df = rels_df[rels_df["ismain"]].drop("ismain", axis=1)


rels_df
```

```python
import ast
def separar(row):
    lista = ast.literal_eval(row["sorted_row"])
    row["ch1"] = lista[0]
    row["ch2"] = lista[1]
    return row
```

```python
rels_df['sorted_row'] = [sorted([a,b]) for a,b in zip(rels_df.ch1, rels_df.ch2)]
rels_df['sorted_row'] = rels_df['sorted_row'].astype(str)
rels_count_df = rels_df.groupby(['sorted_row']).size().reset_index().rename(columns={0:"cuenta"})
rels_count_df = rels_count_df.apply(separar, axis=1)
rels_count_df = rels_count_df.drop("sorted_row", axis=1).sort_values("cuenta", ascending=False)
rels_count_df["cuenta"] = rels_count_df["cuenta"]**2
rels_count_df
```

```python
dict_ch = {"Chandler":0, "Joey":1, "Ross":2, "Monica":3, "Rachel":4, "Phoebe":5}
rels_count_df["ch1"] = rels_count_df["ch1"].map(dict_ch)
rels_count_df["ch2"] = rels_count_df["ch2"].map(dict_ch)
rels_count_df = rels_count_df.sort_values("cuenta")
rels_count_df
```

```python
df_nodes = pd.DataFrame(pd.Series(dict_ch)).reset_index().rename(columns={0:"index", "index":"name"})
nodes = hv.Dataset(df_nodes, 'index')
nodes.data
```

```python
chord = hv.Chord((rels_count_df, nodes), ['ch1', 'ch2'], ['cuenta'])

paleta = [
    "#243B48",
    "#F49B4C",
    "#3C976F",
    "#839DA5",
    "#090B08",
    "#265594"]


chord.opts(
    opts.Chord(cmap="Category10", edge_cmap="Blues", edge_color=dim('cuenta').str(), 
             node_color=dim('index').str(),edge_alpha=1,node_size=40, edge_linewidth=1.1,
              bgcolor="#F5EABC"))
```

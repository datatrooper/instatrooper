```python
with open("../../../data/sw/starwars-episode-7-interactions.json") as json_file:
    data = json.load(json_file)
```

```python
data["nodes"]
```

```python
df_nodes = pd.DataFrame.from_dict(data["nodes"]).reset_index().rename(columns={"index":"indice"})
df_nodes2 = df_nodes[df_nodes["colour"] != "#808080"]
importantes = df_nodes2["indice"].values
```

```python
def get_important(row):
    if ((row["source"] in importantes) & (row["target"] in importantes)):
        row["important"] = 1
    else:
        row["important"] = 0
    return row
```

```python
links = pd.DataFrame.from_dict(data["links"])
links = links.apply(get_important, axis=1)
links = links[links["important"] == 1]
links = links.drop("important", axis=1)
links.sort_values("value")
```

```python
total_part = list(links["source"].values)
total_part.extend(list(links["target"].values))
```

```python
df_nodes = df_nodes[df_nodes["indice"].isin(total_part)].drop(["value", "indice"], axis=1)
nodes = hv.Dataset(df_nodes, 'index')
nodes.data
```

```python
plt.style.use('https://raw.githubusercontent.com/datatrooper/instatrooper/master/styles/whitetrooper.mplstyle')

chord = hv.Chord((links, nodes), ['source', 'target'], ['value'])

paleta = [
    "#243B48",
    "#F49B4C",
    "#3C976F",
    "#839DA5",
    "#090B08",
    "#265594"]


chord.opts(
    opts.Chord(edge_cmap="Blues", edge_color=dim('value'), 
             node_color=dim('colour').str(),edge_alpha=1,node_size=30, edge_linewidth=5.1,
              bgcolor="#F5EABC", labels='name'))
```

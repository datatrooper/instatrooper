## Datos Champions League

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../../data/futbol/champions.csv', header=None, engine='python')
```

```python
df.columns = ['date','place', 'winner', 'second','score']
df = df.set_index('date')
df = df[['winner']]
```

```python
df[df['winner']=='R.Madrid'] = 'Real Madrid'
df[df['winner']=='Inter'] = 'Inter de Milán'
df[df['winner']=='Milan'] = 'A.C Milan'
```

```python
df
```

```python
df_camp = pd.pivot_table(df, index=['date'],
                    columns=['winner'], aggfunc=lambda x: np.count_nonzero(x))
```

```python
df_camp
```

```python
df_camp.columns
```

```python
df_camp_ac = df_camp.fillna(0).cumsum()
```

```python
df_camp_ac
```

```python
df_camp_ac.index = pd.to_datetime(df_camp_ac.index, format = '%Y').strftime('%Y')
```

```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt


from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go


sns.set_context("talk")
plt.style.use('ggplot')
mpl.rcParams['axes.labelpad']='20'
```

### data en https://data.worldbank.org/indicator/SP.POP.TOTL.MA.ZS

```python
df = pd.read_csv("../datos/male_population.csv")
```

```python
df.head()
```

```python
import numpy as np
np.round(0.51)
```

```python
years = list(range(1960,2020))
years = [str(x) for x in years]
df_bool = df.copy()
df_bool[years] = df[years].apply(lambda x: np.round(x/100), axis=1)
df_bool.head()
```

```python
for year in years:

    data=dict(
        type = 'choropleth',
        locations = df_bool['Country Code'],
        z = df_bool[year],
        text = df_bool['Country Name'],
        colorscale = 'Blues',
        marker_line_color='darkgray',
        marker_line_width=0.5,
        showscale = False
        
    )

    layout = dict(geo=dict(
            showframe=False,
            showcoastlines=True,
            projection_type='equirectangular'
        ))

    fig = go.Figure(data = [data], layout = layout)

    fig.update_layout(
        title=year,
        font=dict(
            family="Courier New, monospace",
            size=28,
            color="#7f7f7f"
        )
    )

    iplot(fig)
```

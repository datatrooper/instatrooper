```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from stop_words import get_stop_words
from collections import Counter
import random



sns.reset_defaults()

datatrooper_palette = [
    "#E6D3AD",
    "#967B4F",
    "#46493D",
    "#839DA5",
    "#9FADAD",
    "#2B4144",
    "#986E37"
]

plt.style.use('fivethirtyeight')
sns.set_palette(datatrooper_palette)
#  #1D1D3D o black
sns.set(rc={'axes.facecolor':'#1D1D3D', 'figure.facecolor':'#1D1D3D'})

sns.set_context("talk")
COLOR = 'white'
mpl.rcParams['text.color'] = COLOR
mpl.rcParams['axes.labelcolor'] = COLOR
mpl.rcParams['xtick.color'] = COLOR
mpl.rcParams['ytick.color'] = COLOR
```

```python
df = pd.read_csv("../../../data/geoMap.csv")
df['lugar'] = df.index
df = df.drop('Región')
```

```python
df.head(22)
```

```python
plt.style.use('https://raw.githubusercontent.com/datatrooper/instatrooper/master/styles/whitetrooper.mplstyle')
fig = plt.figure(figsize=(5,10))


#sns.set(font_scale = 0.8)


ax = sns.barplot(y="Categoría: Todas las categorías", x='lugar', orient="h", data = df.head(22), palette=datatrooper_palette)
#plt.xlim(8, 8.6)
ax.set_title("2004 - 2018", pad=20, fontsize=35)
ax.set_xlabel("Score", fontsize=27)
ax.set_ylabel("Character", fontsize=27)

plt.show()
fig.savefig('starwars_total.png', facecolor=fig.get_facecolor(),pad_inches=2, bbox_inches="tight", dpi=100)
```

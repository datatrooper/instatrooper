```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from stop_words import get_stop_words
from collections import Counter
import random

sns.reset_defaults()

datatrooper_palette = [
    "#E6D3AD",
    "#967B4F",
    "#46493D",
    "#839DA5",
    "#9FADAD",
    "#2B4144",
    "#986E37"
]

plt.style.use('fivethirtyeight')
sns.set_palette(datatrooper_palette)
#  #1D1D3D o black
sns.set(rc={'axes.facecolor':'#1D1D3D', 'figure.facecolor':'#1D1D3D'})

sns.set_context("talk")
COLOR = 'white'
mpl.rcParams['text.color'] = COLOR
mpl.rcParams['axes.labelcolor'] = COLOR
mpl.rcParams['xtick.color'] = COLOR
mpl.rcParams['ytick.color'] = COLOR
```

```python
df = pd.read_csv("../datos/tumbzilla_labels.csv").drop("Unnamed: 0", axis=1)
```

```python
df["voting"].nunique()
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x="voting", data=df, palette=datatrooper_palette)
plt.xticks(rotation=90)
plt.show()
```

```python
words = df["title"].str.cat(sep=' ').lower().split(" ")

stopwords = get_stop_words('en')
mylist = ["gets", "The", "scene", "-"]
stopwords.extend(mylist)
words = [word for word in words if word not in stopwords]
word_count = Counter(words)
```

```python
word_count
```

```python
def white_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "rgb(230, 217, 168)" 
```

```python
wc = WordCloud(background_color="#1D1D3D",
               width=1000, 
               height=1500).generate_from_frequencies(word_count)
wc.recolor(color_func=white_color_func, random_state=3)

fig = plt.figure(figsize=(15,15))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

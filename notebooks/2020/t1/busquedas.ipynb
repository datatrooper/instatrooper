```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.reset_defaults()

plt.style.use('https://raw.githubusercontent.com/datatrooper/instatrooper/master/styles/bluetrooper.mplstyle')
datatrooper_palette = [
    "#e6d3ad",
    "#967B4F",
    "#46493D",
    "#839DA5",
    "#9FADAD",
    "#2B4144",
    "#986E37"
]
```

```python
df = pd.read_csv("../../../data/busquedas/enfermedades.csv")
df = df.drop("Region")
df = df.reset_index().rename(columns={"index": "dpto", "Category: All categories": "busquedas"})
df["busquedas"] = df["busquedas"].fillna(0).astype(int)
df = df.sort_values("busquedas", ascending=False).head(5)
```

```python
ax = sns.barplot(x="busquedas", y="dpto", data=df, palette = datatrooper_palette)
plt.grid(False)
ax.set_xlabel("", fontsize=30)
plt.xticks([])
plt.box(False)
ax.set_ylabel("", fontsize=30)
ax.tick_params(axis='both', which='major', labelsize=50)

for index, row in df.iterrows():
    ax.text(row["busquedas"] + 6,index+0.1, row["busquedas"], color='white', ha="center", size=35)

plt.savefig('match.png',bbox_inches='tight', transparent="True", pad_inches = 0)
plt.show()
```
